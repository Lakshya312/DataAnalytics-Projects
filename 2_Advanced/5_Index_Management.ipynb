```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(5)
```

```python
df.index
```

```python
df.index.name = 'job_index'
```

```python
df.head()
```

```python
df.index.dtype
```

# Three Methods to Cover
* df.reset_index()
* df.set_index()
* df.sort_index()

# reset_index()

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa.head()
```

```python
df_usa.reset_index(inplace=True)
df_usa
```

# set_index()

```python
df_usa.set_index('job_index', inplace=True)
df_usa
```

# sort_index()

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'mean', 'std', 'max', 'min'])
median_pivot
```

```python
median_pivot.sort_values(by='job_title_short', ascending=True)
```

```python
# now we can use sort_index on median_pivot table
median_pivot.sort_index(inplace=True)
median_pivot
```

# 🟩 Set Index and Display Rows (2.5.1)

```python
df_india = df[df['job_country'] == 'India']
```

```python
df_india.index.name = 'job_index'
df_india
```

# 🟩 Reset Index and Drop Existing (2.5.2)

```python
df_india.reset_index(drop=True, inplace=True)

df_india
```

# 🟩 Sort Index (2.5.3)

```python
mean_pivot = df_india.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'mean', 'max'])
mean_pivot.sort_values(by=('mean', 'salary_year_avg'), ascending=False)
```

# 🟨 Pivot Table Reset & Set Index (2.5.4)

```python
mean_pivot.reset_index(inplace=True)
mean_pivot
```

```python
mean_pivot.set_index('job_title_short', inplace=True)
mean_pivot
```
