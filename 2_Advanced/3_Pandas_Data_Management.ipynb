```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_altered
```

```python
median_year_salary = df_altered['salary_year_avg'].median()
median_hour_salary = df_altered['salary_hour_avg'].median()
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_year_salary)
df_altered['salary_hour_avg'] = df_altered['salary_hour_avg'].fillna(median_hour_salary)
```

# 🟩 Set Index and Display Rows (2.5.1)

```python
df_altered = df_altered.set_index('job_title')
df_altered.iloc[0:7, 0:6]
```

# 🟩 Reset Index and Drop Existing (2.5.2)

```python
df_altered  = df_altered.reset_index(drop=True)#drop=true removes the old index column completely i.e job_tiltle column
df_altered
```

# 🟩 Sort Index (2.5.3)

```python
#df_altered.sort_index(ascending=False, inplace=True)
df_altered.sort_index(inplace=True)
df_altered
```
