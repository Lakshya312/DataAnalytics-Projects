```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.head()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
df.head()
```

```python
months = df['job_posted_month'].unique()

# Now building dictionary comprehension
'''x = {}
for month in months:
    x[month] = month
x'''
dict_months = {month:df[df['job_posted_month'] == month] for month in months}
dict_months['Jan']
```

# BOSS NEEDS 2023 QUARTER 1 

```python
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
df_q1
```

```python
# checking weather they are concatinated or not
df_q1['job_posted_month'].value_counts().plot(kind='bar')
```

```python
# making a unique list of all the job postings
job_posting = df['job_title_short'].unique()

# distributing the data with respect to job posting as dictionaries with dictionary comprehension
dict_job_posting = {job:df[df['job_title_short'] == job] for job in job_posting}
```

# Concatinating top 3 Job postings

```python
top_3 = df['job_title_short'].value_counts().head(3).index
df_top_3  = pd.concat([dict_job_posting[title] for title in top_3], ignore_index=True)
```

```python
df_top_3['job_title_short'].value_counts().plot(kind='bar')
```
