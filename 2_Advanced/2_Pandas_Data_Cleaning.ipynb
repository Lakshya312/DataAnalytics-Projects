```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
median_year_avg = df['salary_year_avg'].median()
median_year_avg
```

```python
median_hour_avg = df['salary_hour_avg'].median()
median_hour_avg
```

```python
df_filled  = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_year_avg)#nan values inside median_year_avg have been replaced by median values 
df['salary_hour_avg'] = df['salary_hour_avg'].fillna(median_hour_avg)
```

```python
df_filled.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
df_unique = df_filled
df_unique = df_unique.drop_duplicates()
```

```python
print('Length of original df:           ', len(df_filled))
print('Length of drop duplicates df:    ', len(df_unique))
print('Rows dropped                     ', len(df_filled) - len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])
df_unique
```

```python
print('Length of original df:           ', len(df_filled))
print('Length of drop duplicates df:    ', len(df_unique))
print('Rows dropped                     ', len(df_filled) - len(df_unique))
```
