```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df
```

```python
df.describe()
```

```python
df['salary_year_avg'].median()
```

```python
df['salary_year_avg'].min()
```

```python
min_salary = df['salary_year_avg'].idxmin()
```

```python
df.iloc[min_salary]
```

```python
df.groupby('job_title_short')['job_title_short'].min()
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
grouped = df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
print(grouped)
```

```python
grouped = df.groupby(['job_title_short', 'job_country'])[['salary_year_avg', 'salary_hour_avg']].median()
print(grouped)
```

```python
grouped = df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].agg(['min', 'max', 'median'])
print(grouped)
```
