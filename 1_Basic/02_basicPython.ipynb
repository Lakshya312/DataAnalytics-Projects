#Classes: The templates of objects; it's a way to generate objects

```python
job_title = "Data Analyst"
job_location = "United States"
job_salary = 9000
```

```python
class JobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
```

```python
JobPost(job_title, job_location, job_salary)
```

#Attributes: The variables of an objects

```python
job_1 = JobPost(job_title, job_location, job_salary)
```

```python
job_1.title
```

# Methods = functions of an objects

```python
class JobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
  def print_job_title(self):
    print(self.title)
```

```python
job_4 = JobPost(job_title, job_location, job_salary)
```

```python
job_4.print_job_title()
```
