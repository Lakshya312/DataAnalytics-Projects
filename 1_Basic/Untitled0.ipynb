```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt # we will use this for plotting today

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#cleaning data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df
```

```python
randomDate = pd.DataFrame({
    'year': [2020, 2021, 2022, 2023],
    'salary': [50000, 55000, 60000, 70000]
})
plt.plot(randomDate['year'], randomDate['salary'])
plt.show()
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind='barh')
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.title('Median Salary by job')
plt.show()
```
