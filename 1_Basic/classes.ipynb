```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate = 0.1, symbol = '$'):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self. total_salary = base_salary + (base_salary * bonus_rate)

  def __repr__(self):
    return f"{self.symbol}{self.base_salary:,.2f}"

  def calculate_salary(self):
    return f"{self.symbol}{self.total_salary:,.2f}"

  def calculate_bonus(self):
    return f"{self.symbol}{(self.total_salary - self.base_salary):,.2f}"
```

```python
salary = BaseSalary(100000)
salary.symbol
```
