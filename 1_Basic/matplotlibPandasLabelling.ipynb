#🟩 Job Posting Location Bar Plot (1.31.1)

```python
job_location = df['job_location'].value_counts().head(10)
job_location = job_location.sort_values(ascending=False)
job_location
```

```python
job_location.plot(kind='bar')
plt.title('Postings by job location')
plt.ylabel('Count of job postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#🟩 Top Job Locations (1.31.2)

```python
job_location.head(5).plot(kind='bar')
plt.title('Top 5 postings by job location')
plt.ylabel('Count of job postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```
